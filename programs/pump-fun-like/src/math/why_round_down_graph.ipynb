```python
import numpy as np
import matplotlib.pyplot as plt

# 定义参数范围
Y = np.linspace(0, 0.00000001, 10)  # Y 的范围，最小为1
x = 1  # x 最小为1
# x = np.linspace(1, 100, 500)

# 定义公式
def delta_x_formula(Y, x, offset=0):
    numerator = (Y + offset) * 1000000073
    denominator = 30 + (Y + offset)
    return numerator / denominator - x

# 计算 Δx 的两条曲线
delta_x_1 = delta_x_formula(Y, x, offset=0)
delta_x_2 = delta_x_formula(Y, x, offset=0.000000001)

# 绘制曲线
plt.figure(figsize=(10, 6))
plt.plot(Y, delta_x_1, label=r"$\Delta x = \frac{Y \cdot 1073^{12}}{30^9 + Y} - x$", color='blue')
plt.plot(Y, delta_x_2, label=r"$\Delta x = \frac{(Y+1) \cdot 1073^{12}}{30^9 + (Y+1)} - x$", color='orange')

# 设置图例和标签
plt.title("Mixed Curve of Δx Formulas", fontsize=14)
plt.xlabel("Y", fontsize=12)
plt.ylabel(r"$\Delta x$", fontsize=12)
plt.legend(fontsize=12)
plt.grid(alpha=0.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 定义参数范围
Y = np.linspace(0, 0.00000001, 10)  # Y 的范围，最小为1
x = 1  # x 最小为1
# x = np.linspace(1, 100, 500)

# 定义公式
def delta_x_formula(Y, x, offset=0):
    numerator = (Y + offset) * 1000000073
    denominator = 30 + (Y + offset)
    return x - numerator / denominator

# 计算 Δx 的两条曲线
delta_x_1 = delta_x_formula(Y, x, offset=0)
delta_x_2 = delta_x_formula(Y, x, offset=0.000000001)

# 绘制曲线
plt.figure(figsize=(10, 6))
plt.plot(Y, delta_x_1, label=r"$\Delta x = x - \frac{Y \cdot 1073^{12}}{30^9 + Y}$", color='blue')
plt.plot(Y, delta_x_2, label=r"$\Delta x = x - \frac{(Y+1) \cdot 1073^{12}}{30^9 + (Y+1)}$", color='orange')

# 设置图例和标签
plt.title("Mixed Curve of Δx Formulas", fontsize=14)
plt.xlabel("Y", fontsize=12)
plt.ylabel(r"$\Delta x$", fontsize=12)
plt.legend(fontsize=12)
plt.grid(alpha=0.5)
plt.show()
```
